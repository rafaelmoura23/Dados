**Solu√ß√£o do Problema de encontrar a maior idade utilizando Listas**

```python
# lista para armazenar 3 idades
#idades=[27, 49, 12]

# lista para armazenar 10 idades
idades=[27, 49, 12, 67, 21, 32, 18, 45, 84, 53, 22, 56, 80, 35, 18]

# inicialmente assume que a primeira idade √© a maior
maior_idade = idades[0]

# percorre a lista verificando a maior idade
for idade in idades:
  if idade > maior_idade:
    maior_idade = idade

# exibe a maior idade encontrada
print ('Maior idade: ', maior_idade)

# m√©todo MAX (retorna o maior valor de uma lista)
print('Maior idade: ', max(idades))
```

```python
# Cria√ß√£o dos conjuntos A e B
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}
print('A:', A)
print('B:', B)
```

```python
# Opera√ß√£o de Uni√£o: (A ‚à™ B)
print('A | B =>', A | B)
print('A.union(B) =>', A.union(B))

# retorna todos os elementos de A e B e aglutina os repetidos
```

```python
# Opera√ß√£o de Interse√ß√£o: (A ‚à© B)
print('A & B =>', A & B)
print('A.intersection(B) =>', A.intersection(B))

# retorna os elementos que se encontram em A e B
```

```python
# Opera√ß√£o de Diferen√ßa: (A - B) e (B - A)
print('A - B =>', A - B)
print('A.difference(B) =>', A.difference(B))
# tudo o que est√° em A mas n√£o est√° em B
print('B - A =>', B - A)
print('B.difference(A) =>', B.difference(A))
# tudo o que est√° em B mas n√£o est√° em A
```

```python
# Cria√ß√£o dos conjuntos A e B
c1 = {1, 2, 3, 4, 5}
c2 = {4, 5}
c3 = {91, 92, 93}
```

```python
# Adiciona um elemento ao conjunto
c1.add(6)
print(c1)
```

```python
# Adiciona os elementos de uma sequ√™ncia iter√°vel
c1.update([2, 4, 6, 8])
print(c1)
```

```python
# Descarta um elemento do conjunto,
c1.discard(8)
print(c1)

# Diferentemente do set.remove(), o discard n√£o gera um erro # se o elmento a ser removido n√£o existir
c1.discard(99)
print(c1)
```

```python
# Verifica se os conjuntos s√£o disjuntos, ou seja,
# se n√£o possuem nenhum elemento em comum
print(c1.isdisjoint(c2))
print(c1.isdisjoint(c3))
```

```python
# Verifica se o conjunto √© subconjunto de outro
print(c1.issubset(c2))
print(c2.issubset(c1))
```

```python
# Verifica se o conjunto cont√©m outro conjunto (superset)
print(c1.issuperset(c2))
print(c2.issuperset(c1))
```

```python
# Cria o dicion√°rio
d1 = {'zero': 0, 'um': 1, 'dois': 2, 'tr√™s': 3, 'quatro': 4}
print(d1)
# Encontra a maior e menor chave
print('Maior e menor chave:', max(d1), min(d1))
# Adiciona elementos de um outro dicionario
d1.update({'cinco': 5, 'seis': 6})
print(d1)
# Verifica se o dicion√°rio possui as seguintes chaves
print("A chave 'dois' est√° no dicion√°rio?", 'dois' in d1)
print("A chave 'cinco' n√£o est√° no dicion√°rio?", 'dois' in d1)
# Remove o elemento com chave 'zero'
d1.pop('zero')
print(d1)
```

```python
# Cria o dicion√°rio
d1 = {'zero': 0, 'um': 1, 'dois': 2, 'tr√™s': 3, 'quatro': 4}
# Itera sobre as chaves
for key in d1:
  if key == 'tr√™s':
    print('Chave tr√™s encontrada!')
for key in d1.keys():
  if key == 'dez':
    print('Chave quatro encontrada!')
# Itera sobre os valores
soma = 0
for value in d1.values():
  soma = soma + value
print('Soma dos valores do dicion√°rio:', soma)
# Itera sobre os itens
for key, value in d1.items():
    print(key, value)
```

# **FUN√á√ïES**

def nome_da_funcao(argumento1, argumento2, ...):
 <bloco de c√≥digo a ser executado>
 return resultado

Sua sintaxe √© formada pelos seguintes componentes:
‚óè A palavra-chave def, que define o in√≠cio da declara√ß√£o de uma
fun√ß√£o.
‚óè O identificador da fun√ß√£o (nome_da_funcao), que deve ser
√∫nico e que ser√° utilizado para chamar esta fun√ß√£o. Os nomes
de fun√ß√µes devem seguir as mesmas regras dos nomes de
vari√°veis, apresentados na Se√ß√£o Regras para Nomea√ß√£o de
Vari√°veis.
‚óè Os argumentos (argumento1, argumento2, ...), que s√£o
opcionais, s√£o utilizados para repassar valores paras as
fun√ß√µes. Mesmo que n√£o for passado nenhum argumento
para a fun√ß√£o, √© necess√°rio declarar os par√™nteses
(nome_da_funcao()).
‚óè O tradicional dois pontos (:) para marcar o fim da declara√ß√£o e
o in√≠cio de um bloco indentado.
104
‚óè O <bloco de c√≥digo>, que ir√° compor o corpo da fun√ß√£o, √© o
conjunto de instru√ß√µes respons√°vel por realizar as
computa√ß√µes que a fun√ß√£o ir√° desempenhar.
‚óè Por fim, uma instru√ß√£o de retorno (return), que √© respons√°vel
por retornar o resultado da fun√ß√£o. Uma vez que existem
fun√ß√µes que n√£o retornam valores, sua declara√ß√£o √© opcional.

```python
def divisor_invalido(divisor):
  if divisor==0:
    print('ERRO: Divisor igual a zero!')
    return True
  return False


def div(dividendo, divisor):
  if divisor_invalido(divisor):
      return
  return dividendo / divisor
```
